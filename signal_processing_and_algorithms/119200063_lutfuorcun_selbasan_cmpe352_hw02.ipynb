```python
from thinkdsp import read_wave, decorate

firstWave = read_wave('119200063_lutfuorcun_selbasan_cmpe352_hw02_sentence_01.wav')
secondWave = read_wave('119200063_lutfuorcun_selbasan_cmpe352_hw02_sentence_02.wav')
```

```python
  # Audio of first sentence
firstWave.make_audio()
```

```python
# Audio of second sentence
secondWave.make_audio()
```

```python
# A1.1) Waveform of first sentence
firstWave.plot()
```

```python
# A1.2) Spectrogram of first sentence
firstWaveSpectrogram = firstWave.make_spectrogram(256)
firstWaveSpectrogram.plot(high=3000)
```

```python
# A1.3) Spectrum of first sentence
firstSpectrum = firstWave.make_spectrum()
firstSpectrum.plot()
```

```python
# A2.1) Applying different window functions to first copy of first sentence
firstWave01 = firstWave
for window_func in [np.bartlett, np.blackman, np.hamming, np.hanning]:
    firstWave01.ys *= window_func(len(firstWave01.ys))

    spectrum = firstWave01.make_spectrum()
    spectrum.plot(high=1000, label=window_func.__name__)
```

```python
# A3.1) Applying hamming window function to second copy of first sentence
firstWave02 = firstWave
window02 = np.hamming(len(firstWave02))
firstWave02.window(window02)
firstWave02.plot()
# Starting and ending ys is significantly reduced compared to original first sentence
```

```python
# A3.1) Audio after hamming function
firstWave02.make_audio() 
```

```python
# A3.1) Spectrogram after hamming function
fW02spectrogram = firstWave02.make_spectrogram(256)
fW02spectrogram.plot(high=2000)
```

```python
# A4.1) Applying bartlett window function to third copy of first sentence
firstWave03 = firstWave
window03 = np.bartlett(len(firstWave03))
firstWave03.window(window03)
firstWave03.plot()
# More or less what happened is same as hamming, but bartlet reduced more
```

```python
# A4.1) Spectrogram of third copy after bartlett
fW03spectrogram = firstWave03.make_spectrogram(256)
fW03spectrogram.plot(high=2000)
```
